# Regplot 
- https://seaborn.pydata.org/generated/seaborn.regplot.html

```python
from sklearn.datasets import load_boston
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

dataset = load_boston()
feature = dataset.data
DF = pd.DataFrame(feature,columns=dataset.feature_names)
DF['price'] = dataset.target
```

## 회귀직선 포함 산점도
- price와 나머지 열과의 관계를 확인해보자.

```python
fig, axs = plt.subplots(figsize = (20,10), ncols=5, nrows=3,constrained_layout = True)
lm_features = DF.iloc[:,:-1]
for i, feature in enumerate(lm_features):
    row = int(i/5)
    col = i%5
    sns.regplot(x=feature , y='price', marker='.', data=DF, ax=axs[row][col], line_kws={'color': 'red'}, scatter_kws={'color': 'darkblue'})
```

## 산점도 
- `fit_reg = False`
- price와 나머지 열과의 관계를 확인해보자.

```python
fig, axs = plt.subplots(figsize = (20,10), ncols=5, nrows=3,constrained_layout = True)
lm_features = DF.iloc[:,:-1]
for i, feature in enumerate(lm_features):
    row = int(i/5)
    col = i%5
    sns.regplot(x=feature , y='price', marker='.', data=DF, ax=axs[row][col],line_kws={'color': 'red'}, scatter_kws={'color': 'darkblue'}, fit_reg=False)
```
