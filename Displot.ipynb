# Displot 
- `Density`, `CDF`, `histogram` 
- https://seaborn.pydata.org/generated/seaborn.displot.html

```python
import seaborn as sns, numpy as np
import matplotlib.pyplot as plt

penguins = sns.load_dataset('penguins')

# select the columns to be plotted
cols = ['bill_length_mm', 'bill_depth_mm']

# create the figure and axes
fig, axes = plt.subplots(1, 2)
axes = axes.ravel()  # flattening the array makes indexing easier

for col, ax in zip(cols, axes):
    sns.histplot(data=penguins[col], kde=True, stat='density', ax=ax)

fig.tight_layout()
plt.show()
```

```python
sns.displot(data=penguins, x="flipper_length_mm", kde=True)
sns.displot(penguins["flipper_length_mm"],kind='kde')
sns.displot(data=penguins, x="flipper_length_mm", kind="ecdf")
```
